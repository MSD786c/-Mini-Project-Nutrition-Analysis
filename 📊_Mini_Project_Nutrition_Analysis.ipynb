## Boxplots


food categories boxplots to visualize the distribution of Calories across these categories and identify outliers.


```python
def categorize_food(food_name):
    food_name = food_name.lower()
    if 'cheese' in food_name or 'milk' in food_name or 'yogurt' in food_name:
        return 'Dairy'
    elif 'chicken' in food_name or 'beef' in food_name or 'pork' in food_name or 'lamb' in food_name:
        return 'Meat'
    elif 'salmon' in food_name or 'tuna' in food_name or 'shrimp' in food_name or 'fish' in food_name:
        return 'Seafood'
    elif 'rice' in food_name or 'bread' in food_name or 'pasta' in food_name or 'oats' in food_name:
        return 'Grains'
    elif 'apple' in food_name or 'banana' in food_name or 'berry' in food_name or 'orange' in food_name:
        return 'Fruit'
    elif 'brocoli' in food_name or 'spinach' in food_name or 'carrot' in food_name or 'potato' in food_name:
        return 'Vegetable'
    elif 'nut' in food_name or 'seed' in food_name:
        return 'Nuts and Seeds'
    elif 'oil' in food_name or 'butter' in food_name or 'margarine' in food_name:
        return 'Fats and Oils'
    elif 'sauce' in food_name or 'dressing' in food_name or 'soup' in food_name:
        return 'Sauces and Soups'
    elif 'burger' in food_name or 'sandwich' in food_name or 'pizza' in food_name:
        return 'Fast Food'
    else:
        return 'Other'

df['food_category'] = df['food'].apply(categorize_food)

plt.figure(figsize=(15, 8))
sns.boxplot(x='food_category', y='Caloric Value', data=df)
plt.title("Distribution of Caloric Value Across Food Categories")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Presentation of Boxplot

print("\n--- Boxplot ---")

# Display the boxplot (assuming it is still in the notebook's output)
# If not, the code to generate it would be here again.

print("\nInsights from Boxplot:")
print("\nDistribution of Caloric Value Across Food Categories:")
print("- The boxplot reveals significant variation in the distribution of caloric values across different food categories.")
print("- 'Fast Food' and 'Meat' categories appear to have higher median caloric values and wider ranges compared to categories like 'Fruit' and 'Vegetable'.")
print("- Many categories show outliers with very high caloric values, indicating specific food items within those categories that are exceptionally calorie-dense.")
print("- Categories like 'Fruit' and 'Vegetable' generally have lower caloric values with fewer outliers.")
```

## Heatmap


Calculate the correlation matrix between numerical nutrient columns and visualize it using a heatmap.


```python
import matplotlib.pyplot as plt
import seaborn as sns

# Identify numerical columns in the dataframe
numerical_cols = df.select_dtypes(include=['number']).columns.tolist()

# Define a list of potential nutrient columns (excluding calculated ones and identifiers)
potential_nutrient_cols = [
    'Caloric Value', 'Fat', 'Saturated Fats', 'Monounsaturated Fats',
    'Polyunsaturated Fats', 'Carbohydrates', 'Sugars', 'Protein',
    'Dietary Fiber', 'Cholesterol', 'Sodium', 'Calcium', 'Copper',
    'Iron', 'Magnesium', 'Manganese', 'Phosphorus', 'Potassium',
    'Selenium', 'Zinc', 'Vitamin A', 'Vitamin C', 'Vitamin D', 'Vitamin E',
    'Vitamin K', 'Thiamin', 'Riboflavin', 'Niacin', 'Pantothenic acid',
    'Vitamin B6', 'Folate', 'Vitamin B12'
]

# Filter the potential nutrient columns to include only those present in the dataframe and are numerical
nutrient_cols_present = [col for col in potential_nutrient_cols if col in numerical_cols]

# Calculate the correlation matrix for the selected numerical nutrient columns
correlation_matrix = df[nutrient_cols_present].corr()

# Create a heatmap of the correlation matrix
plt.figure(figsize=(16, 12))
sns.heatmap(correlation_matrix, annot=False, cmap='coolwarm')
plt.title("Correlation Matrix of Nutrients")
plt.show()

# Presentation of Heatmap

print("\n--- Heatmap ---")

# Display the heatmap (assuming it is still in the notebook's output)
# If not, the code to generate it would be here again.

print("\nInsights from Heatmap:")
print("\nCorrelation Matrix of Nutrients:")
print("- The heatmap visually represents the correlations between different numerical nutrient columns.")
print("- Darker shades of red indicate strong positive correlations, while darker shades of blue indicate strong negative correlations.")
print("- Light shades or white indicate weak correlations.")
print("- Notable strong positive correlations exist between:")
print("  - Caloric Value and Fat")
print("  - Fat and Saturated Fats, Monounsaturated Fats, and Polyunsaturated Fats")
print("  - Carbohydrates and Sugars")
print("  - Protein and various vitamins and minerals (e.g., Vitamin B6, Potassium, Phosphorus)")
print("- Some nutrients show weak or negligible correlations with many others, suggesting they are not strongly related in this dataset.")
print("- The diagonal line represents the correlation of a nutrient with itself, which is always 1.")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Sort by Protein Density and get top/bottom 10
df_sorted_protein_density = df.sort_values(by='Protein Density', ascending=False)
top_10_protein_density = df_sorted_protein_density.head(10)
bottom_10_protein_density = df_sorted_protein_density.tail(10)

# Combine for plotting
protein_density_plot_data = pd.concat([top_10_protein_density, bottom_10_protein_density])

# Create the bar plot
plt.figure(figsize=(12, 8))
sns.barplot(x='food', y='Protein Density', data=protein_density_plot_data)
plt.title('Top 10 Highest and Lowest Protein Density Foods')
plt.xlabel('Food')
plt.ylabel('Protein Density (Protein per Calorie)')
plt.xticks(rotation=90, ha='right')
plt.tight_layout()
plt.show()
```

### Protein Density Bar Plot

This bar plot displays the foods with the highest and lowest protein density. Protein density is calculated as the amount of protein per calorie. Foods with a higher protein density provide more protein relative to their caloric value.

**Insights:**

*   The plot highlights foods that are very protein-rich for their caloric content, which can be useful for those looking to maximize protein intake without excessive calories.
*   Conversely, it shows foods with very low protein density, indicating they provide relatively few grams of protein per calorie.
*   Examining the food names helps identify specific items that are either excellent or poor sources of protein when considering caloric intake.
